```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading and Importing flight data sets
data=pd.read_csv('/flights.csv')
```

```python
data.head()
```

```python
#Note: Copying of data is good practice becasue if you work on live data you are not allow
#to change in real dataset thats why copying of data is very good practice 
df=data.copy()
```

```python
#Going to jump on plot we are observe here that passengers data are dependable on years see when
#year goes to up passengers increases while decreases

ax=sns.lineplot(x='year',y='passengers',data=df,hue=df['month'])
plt.figure(figsize=(10,10))
plt.show()
```

```python
#here we want data between 1950 to 1955 thats our task so we are doing conditional statments 
df_passenger_50_55=df.loc[(df['year']>=1950) & (df['year']<=1955)]
```

```python
from matplotlib.colors import Colormap
```

```python
sns.lineplot(x='year',y='passengers',data=df_passenger_50_55,
             palette=sns.color_palette('Paired',90),
             markers=True,style='month')
plt.xlim(1950,1960)
plt.show()
```

```python
df_pivot=df.pivot('year','month','passengers')
```

```python
sns.lineplot(data=df_pivot)
```
